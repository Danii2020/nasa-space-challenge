```python
import xmitgcm.llcreader as llcreader
%matplotlib inline
import holoviews as hv
from holoviews.operation.datashader import regrid
import matplotlib.pyplot as plt

hv.extension('bokeh')
```

```python
# We select an model from which we will obtain the simulated data
model = llcreader.ECCOPortalLLC2160Model()
```

```python
#We retrieved data from the selected model. Given the massive size of the data, reaching terabytes, we focused solely on a specific aspect:
#the temperature ("Theta") of the oceans.
# In addition, we only focus on the "Lat Lon"..
ds_sst = model.get_dataset(varnames=['Theta'], k_levels=[0], type='latlon')
```

```python
data_t0 = ds_sst.Theta.isel(k=0,  i=slice(7000,8000), j=slice(3500,4500), time=0)
data_t0.plot()
plt.title("Time: " + str(data_t0.time.values))
plt.xticks([])
plt.yticks([])
```

```python
data_t2 = ds_sst.Theta.isel(k=0, i=slice(7000,8000), j=slice(3500,4500), time=5000)
data_t2.plot()
plt.title("Time: " + str(data_t2.time.values))
plt.xticks([])
plt.yticks([])
```

```python
# This is an interactive plot
dataset = hv.Dataset(ds_sst.Theta.isel(k=0, i=slice(7000,8000), j=slice(3500,4500)).astype('f4'))
hv_im = (dataset.to(hv.Image, ['i', 'j'], dynamic=True).options(cmap='Magma', width=950, height=600, colorbar=True))
%output holomap='scrubber' fps=3
regrid(hv_im, precompute=True)
```
